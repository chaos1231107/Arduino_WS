```python
import librosa
import librosa.display

import IPython.display as ipd

import numpy as np
import matplotlib.pyplot as plt

audio_path = path + '/가천대/동아리 활동/audio_file2.wav'
```

```python
baseband,sr = librosa.load(audio_path, sr=100) #sr = samples per second, sampling_rate = 100
t = np.arange(0,19602,1)
plt.plot(t,baseband)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
# print(len(y))
# print(len(t))
```

```python
f_c = 107.7 * (10 ** 6) # carrier ferquency
x_t = [] #modulated signal

for dt in range(len(baseband)):
    carrier_signal = np.sin(2*np.pi*f_c*dt)
    x_t_temp = baseband[dt] * carrier_signal  # modulated signal
    x_t.append(x_t_temp)

plt.plot(t,x_t)
plt.show()
```
